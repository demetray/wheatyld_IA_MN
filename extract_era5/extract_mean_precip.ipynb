```python
import xarray as xr
import os 
import pandas as pd
```

```python
# function for rounding to nearest 1/4
def x_rnd(x):
    return round(x*4)/4
```

```python
# path to annual wheat yields data
path = '/Users/demetrayancopoulos/Desktop/SML310/final_project/gro_ylds/wheatf_mn_ia.xlsx'
y = pd.read_excel(path)
```

```python
# path to monthly precip data
path = '/Users/demetrayancopoulos/Desktop/SML310/final_project/era5/mean_tot_precip_monthly/era5.mean_total_precipitation_rate.monthly.1979-2020.nc'
ds = xr.open_dataset(path)
```

```python
# store monthly total precipitation at t/lat/lon where yield data is recorded
cols = ['yr', 'lat', 'lon', 'jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
precip = pd.DataFrame(index = range(0, len(y.index)), columns=cols)
```

```python
for ix in y.index:
    
    lat = y.loc[ix].lat     # retrieve latitude at this entry
    lat = x_rnd(lat)       # round lat coord to nearest 0.25
    
    lon = y.loc[ix].lon     # retrieve longitude at this entry
    lon = x_rnd(lon)        # round lon coord to nearest 0.25
    
    yr = y.loc[ix].StartDate[0] + y.loc[ix].StartDate[1] + y.loc[ix].StartDate[2] + y.loc[ix].StartDate[3]
    
    # t2m at this lat/lon
    mtpr = ds.mtpr.sel(latitude=lat, longitude=lon)
    
    # store monthly max t2m for this lat/lon/yr
    for i in range(1, 13):
        t = yr + '-' + str(i)
        precip.iat[ix, i+2] = mtpr.sel(time=t).values[0]
    
    # store yr for this point
    col = 0
    precip.iat[ix, col] = yr
    
    # store latitude for this point
    col = 1
    precip.iat[ix, col] = lat
    
    # store longitude for this point
    col = 2
    precip.iat[ix, col] = lon

#spath = '/Users/demetrayancopoulos/Desktop/SML310/final_project/monthly_precip_rate.xlsx'
#precip.to_excel(spath)

precip
```
